```python
import pandas as pd
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
titanic=pd.read_csv("/content/test.csv")
titanic.head()
```

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

# 1. Scatterplot(numerical-numerical)


```python
sns.scatterplot(x='total_bill', y='tip',data=tips)
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

# 2.BarPlot(numerical-categorical)

```python
sns.barplot(x='Pclass', y='Age', data=titanic)
```

```python
sns.barplot(x='Pclass',y='Age',data=titanic,hue=titanic['Sex'])
```

# 3.BoxPlot(numerical-categorical)

```python
sns.boxplot(x='Sex',y='Age',data=titanic,hue=titanic['SibSp'])
```

# 4.Distplot(numerical-categorical)

```python
sns.distplot(titanic[titanic['Sex']=="male"]['Age'],hist=False)
sns.distplot(titanic[titanic['Sex']=="female"]['Age'],hist=False)
```

# 5.HeatMap(categorical-categorical)

```python
iris.head()
```

```python
sns.heatmap(pd.crosstab(iris['species'],iris['sepal_length']))
```

```python
(iris.groupby('species').mean()['petal_length']).plot(kind='bar')
```

# 6.ClusterMap(categorical-categorical)

```python
sns.clustermap(pd.crosstab(iris['species'],iris['petal_length']))
```

# 7.Pairplot

```python
sns.pairplot(iris,hue='species')
```

# 8.Lineplot(numerical-numerical)

```python
sns.boxplot(flights.pivot_table(values='passengers',index='month',columns='year'))
```

```python
sns.clustermap(flights.pivot_table(values='passengers',index='month',columns='year'))
```
